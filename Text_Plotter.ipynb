```python
#Create the graph 

def plot_data(files,color):

  #create dataframe to add data to
  df = pd.DataFrame()

  #Iterate through each files passed to function
  for file_name in files:

    #Specify the path to the file
    file_path = 'your_file_path' + file_name #'your_file_path' can be a folder that contains all the files you want to add to the graph for comparison

    #Open the file for reading
    with open(file_path, 'r') as file:
      #Read the contents of the file
      file_contents = file.read()

    """Once we have accesesed our data, we can now visualize it"""

    x_values = []#Create empty list than we can add values to and visualize as x-axis
    y_values = []#Create empty list than we can add values to and visualize as y-axis

    #initialize counter
    indexcounter = 0


    for line in file_contents.splitlines(): #Separates each line in the document into a string
        if 5 <= indexcounter <= 8900: #Checks if the index of the current line is greater than or equal to the line at which relevant data begins
            parts = line.split('  ') #Separates the two sections of each line according to where the double space is
            part2 = parts[1] #Saves the second section of each line as the variable 'part2'
            subparts = part2.split('000000') #Separates 'part 2' into two sections, separated by '000000'
            if len(subparts) == 2:
                #x_values.append(float(subparts[0]))#The first section of the line is regarded as an x value and added to the list of x values, representing a column
                y_values.append(float(subparts[1])) #The second section of the line is regarded as an y value and added to the list of y values, representing a column
        indexcounter += 1 #As we iterate through each line, the index counter increases by 1, keeping track of the index value of the current line

    #Add pressure data from the current file to the dataframe
    df[str(file_name)] = y_values
    #zeroes = [0 for _ in range(2000)]
    #df[str(file_name)] = df[str(file_name)] + zeroes
    print(df)
    print(len(df))


  ax = df.plot()

  #Set title  axis labels

  ax.set_title('Your_Title')





#list of colors for the plots
color = ['blue']

#list of files to plot
files = [
    #file_1,
    #file_2,
    #file_3,
    #...

]


plot_data(files,color)


```
